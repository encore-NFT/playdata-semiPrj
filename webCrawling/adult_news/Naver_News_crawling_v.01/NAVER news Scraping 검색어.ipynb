```python
from bs4 import BeautifulSoup
import urllib.request
from urllib.parse import quote
import pandas as pd


def get_news(query, page_num=10):

  news_df = pd.DataFrame(columns=('news_url', 'news_title', 'news_sub_title', 'news_author', "news_date", "news_article", "news_source"))
  idx = 0

  url_query = quote(query) # quote() 함수는 parameter로 받은 query를 웹 서버나 브라우저가 인식하기 좋은 ascii 값으로 변환(인코딩)하는 함수 값으로, 그 값은 URL에 사용하기에 안전한 값으로 변환한 것이 입니다. 이 경우 사람이 이해하기 힘든 값으로 변환합니다.
  url = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query=" + url_query
  
  for _ in range(0, page_num):

    search_url = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(search_url, 'html.parser')
    links = soup.find_all('div', {'class':'info_group'})

    for link in links:

      # news_source : 신문사
      news_source = link.find('a', {'class':'info press'}).get_text()
      print("news_source: "+news_source)

      # news_url: 기사 원문 URL
      news_url = link.find('a', {'class':'info'}).get('href')
      print("news_url: "+news_url)

      # if (news_url == "#"):
      #   continue
      # else:
      #   news_link = urllib.request.urlopen(news_url).read()
      #   news_html = BeautifulSoup(news_link, 'html.parser')

      #   # news_title: 기사 제목
      #   news_title = news_html.find('h3', {'id':'articleTitle'})
      #   print("news_title: "+str(news_title))

        
      #   # news_date: 날짜
      #   news_date = news_html.find('span', {'class':'t11'})
      #   print("news_date: "+news_date)

      #   # news_article: 기사 내용
      #   news_article = news_html.find('div', {'id':'articleBodyContents'})
      #   print("news_article: "+news_article)

      #   # news_subtitle: 부제목
      #   news_sub_title = news_html.find('strong', {'id':'media_end_summary'})
      #   print("news_sub_title: "+news_sub_title)

      #   # news_author: 기자
      #   if news_html.find('div', {'id':'journalistcard_summary_name'}):
      #     news_author = news_html.find('div', {'id':'journalistcard_summary_name'}).text
      #     news_author = re.sub("[기자]","", news_author).strip()
      #   elif news_html.find('p', {'id':'b_text'}):
      #     news_author = news_html.find('p', {'id':'b_text'}).text
      #     news_author = re.sub("[|기자|인턴|특파원|\s|\d|a-z|(|@|.|)|]","", news_author).strip()
      #   else:
      #     news_author = "null"
      #   print("news_author: "+news_author)


      news_df.loc[idx] = ['news_url', 'news_title', 'news_sub_title', 'news_author', "news_date", "news_article", "news_source"]
      idx += 1
      print('#', end='')
    try:
      next = soup.find('a', {'class':'btn_next'}).get('href')
      url = "https://search.naver.com/search.naver" + next
    except:
       break

  return news_df
```

```python
query = input('검색 질의: ')
news_df = get_news(query, 2)
print('Done')
```
